```sql
create or replace temporary view driver_standings
AS
Select driver_name,
count(1) as total_races,
sum(calculated_points) as total_points,
round(AVG(calculated_points),2) AS avg_points,
RANK() over(order by round(AVG(calculated_points),2) desc) as driver_rank from f1_presentation.calculated_race_results
group by driver_name
HAVING total_races >= 50
order by avg_points desc
```

```sql
Select 
race_year,
driver_name,
count(1) as total_races,
sum(calculated_points) as total_points,
round(AVG(calculated_points),2) AS avg_points from f1_presentation.calculated_race_results
where driver_name in (select driver_name from driver_standings where driver_rank<=10)
group by driver_name,race_year
order by race_year, avg_points desc
```
