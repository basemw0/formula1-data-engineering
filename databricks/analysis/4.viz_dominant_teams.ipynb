```sql
create or replace temporary view team_standings
AS
Select team_name,
count(1) as total_races,
sum(calculated_points) as total_points,
round(AVG(calculated_points),2) AS avg_points,
RANK() over(order by round(AVG(calculated_points),2) desc) as team_rank from f1_presentation.calculated_race_results
group by team_name
HAVING total_races >= 100
order by avg_points desc
```

```sql
Select 
race_year,
team_name,
count(1) as total_races,
sum(calculated_points) as total_points,
round(AVG(calculated_points),2) AS avg_points from f1_presentation.calculated_race_results
where team_name in (select team_name from team_standings where team_rank<=5)
group by team_name,race_year
order by race_year, avg_points desc
```
