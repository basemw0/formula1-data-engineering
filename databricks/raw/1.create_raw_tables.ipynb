```sql
CREATE DATABASE if not exists f1_raw;
```

```sql
Drop table if exists f1_raw.circuits;
CREATE TABLE IF NOT EXISTS f1_raw.circuits(
  circuitID INT,
  circuitRef STRING,
  name STRING,
  location STRING,
  country STRING,
  lat DOUBLE,
  lng DOUBLE,
  alt INT,
  url STRING
)
USING csv
OPTIONS (header "true")
LOCATION "abfss://landing@formula1dl05.dfs.core.windows.net/bronze/circuits.csv";
```

```sql
Drop table if exists f1_raw.races;
Create table if not exists f1_raw.races(
  raceId INT,
  year INT,
  round INT,
  circuitId INT,
  name STRING,
  date DATE,
  time STRING,
  url STRING
)
USING csv
OPTIONS (header "true")
LOCATION "abfss://landing@formula1dl05.dfs.core.windows.net/bronze/races.csv";
```

```sql
create table if not exists f1_raw.constructors(
  constructorId INT,
  constructorRef STRING,
  name STRING,
  nationality STRING,
  url STRING
)
USING json
OPTIONS (header "true")
LOCATION "abfss://landing@formula1dl05.dfs.core.windows.net/bronze/constructors.json";
```

```sql
drop table if exists f1_raw.drivers;
create table if not exists f1_raw.drivers(
  driverId INT,
  driverRef STRING,
  number INT,
  code STRING,
  name STRUCT<forename: STRING, surname: STRING>,
  dob DATE,
  nationality STRING,
  url STRING
)
USING json
OPTIONS (header "true")
LOCATION "abfss://landing@formula1dl05.dfs.core.windows.net/bronze/drivers.json";
select * from f1_raw.drivers
```

```sql
drop table if exists f1_raw.results;
create table if not exists f1_raw.results(
  resultId INT,
  raceId INT,
  driverId INT,
  constructorId INT,
  number INT,
  grid INT,
  position INT,
  positionText STRING,
  positionOrder INT,
  points FLOAT,
  laps INT,
  time STRING,
  milliseconds INT,
  fastestLap INT,
  rank INT,
  fastestLapTime STRING,
  fastestLapSpeed FLOAT,
  statusId INT,
  url STRING
)
USING json
OPTIONS (header "true")
LOCATION "abfss://landing@formula1dl05.dfs.core.windows.net/bronze/results.json";
```

```sql
drop table if exists f1_raw.pit_stops;
create table if not exists f1_raw.pit_stops(
  driverId INT,
  duration STRING,
  lap INT,
  milliseconds INT,
  raceId INT,
  stop INT,
  time STRING
)
USING json
OPTIONS (header "true", multiline true)
LOCATION "abfss://landing@formula1dl05.dfs.core.windows.net/bronze/pit_stops.json";
Select * from f1_raw.pit_stops
```

```sql
drop table if exists f1_raw.lap_times;
create table if not exists f1_raw.lap_times(
  driverId INT,
  lap INT,
  position INT,
  time STRING,
  raceId INT,
  milliseconds INT
)
USING csv
LOCATION "abfss://landing@formula1dl05.dfs.core.windows.net/bronze/lap_times";
select * from f1_raw.lap_times
```

```sql
drop table if exists f1_raw.qualifying;
create table if not exists f1_raw.qualifying(
  qualifyId INT,
  raceId INT,
  driverId INT,
  constructorId INT,
  number INT,
  position INT,
  q1 STRING,
  q2 STRING,
  q3 STRING
)
USING json
OPTIONS (header "true",
multiline true)
location "abfss://landing@formula1dl05.dfs.core.windows.net/bronze/qualifying";
select * from f1_raw.qualifying
```
